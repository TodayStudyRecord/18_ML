```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df['Age'] = df['Age'].fillna(df['Age'].median())      # 나이 결측 채우기
X = pd.get_dummies(df[['Pclass','Sex','Age','SibSp','Parch','Fare']])   # 글자 -> 0/1
y = df['Survived']

X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
model = RandomForestClassifier(n_estimators=200, random_state=0).fit(X_tr, y_tr)
print('정확도 :', round(accuracy_score(y_te, model.predict(X_te)), 3))
print('ROC-AUC:', round(roc_auc_score(y_te, model.predict_proba(X_te)[:, 1]), 3))
```
